```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Convex function: z = x^2 + y^2 (Hessian is positive definite)
Z_convex = X**2 + Y**2

# Concave function: z = -(x^2 + y^2) (Hessian is negative definite)
Z_concave = -(X**2 + Y**2)

# Create plots
fig = plt.figure(figsize=(14, 6))

# Plot Convex Function
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z_convex, cmap='viridis', edgecolor='none')
ax1.set_title('Convex Quadratic Function\n$f(x,y) = x^2 + y^2$')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# Plot Concave Function
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z_concave, cmap='plasma', edgecolor='none')
ax2.set_title('Concave Quadratic Function\n$f(x,y) = -(x^2 + y^2)$')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

plt.tight_layout()
plt.show()
```

When it comes to multi variate the convexity is determined by the Hessian Matrix. If the Hessian is postive its a bowl and convex. If the Hessian is negative its convave. The concavity reduces to convexity since a functions is concave only if the graph is convex. It gives a better understanding of both how concavity and convexity work cause it shows us in a larger scale.
