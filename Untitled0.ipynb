## 1. Convexity/Concavity of Bivariate Quadratic Functions

A general quadratic in two variables can be written as

$$f(x) = \tfrac{1}{2}x^\top Q x + b^\top x + c, \quad x\in\mathbb{R}^2.$$

The Hessian is $\nabla^2 f(x)=Q$. The sign of the eigenvalues of $Q$ determines convexity/concavity:
- **Positive definite** $Q$ $\Rightarrow$ convex (bowl)
- **Negative definite** $Q$ $\Rightarrow$ concave (dome)
- **Indefinite** $Q$ $\Rightarrow$ saddle (neither convex nor concave)

```python
# Grid
x = np.linspace(-3, 3, 120)
y = np.linspace(-3, 3, 120)
X, Y = np.meshgrid(x, y)

# Quadratic forms
Q_convex = np.array([[2.0, 0.2], [0.2, 1.0]])     # positive definite
Q_concave = -Q_convex                              # negative definite
Q_saddle = np.array([[1.0, 0.0], [0.0, -1.5]])      # indefinite

Z_convex = 0.5*(Q_convex[0,0]*X**2 + 2*Q_convex[0,1]*X*Y + Q_convex[1,1]*Y**2)
Z_concave = 0.5*(Q_concave[0,0]*X**2 + 2*Q_concave[0,1]*X*Y + Q_concave[1,1]*Y**2)
Z_saddle = 0.5*(Q_saddle[0,0]*X**2 + 2*Q_saddle[0,1]*X*Y + Q_saddle[1,1]*Y**2)

fig = plt.figure(figsize=(15, 8))

# 3D surfaces
for i, (Z, title) in enumerate([
    (Z_convex, "Convex: positive definite"),
    (Z_concave, "Concave: negative definite"),
    (Z_saddle, "Saddle: indefinite")
]):
    ax = fig.add_subplot(2, 3, i+1, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x,y)')

# Contours
for i, (Z, title) in enumerate([
    (Z_convex, "Convex level sets"),
    (Z_concave, "Concave level sets"),
    (Z_saddle, "Saddle level sets")
]):
    ax = fig.add_subplot(2, 3, i+4)
    ax.contourf(X, Y, Z, levels=20, cmap='viridis')
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.tight_layout()
plt.show()
```

**Why this differs from univariate quadratics:** in one variable, the sign of the single coefficient on $x^2$ determines convex vs concave. In multiple variables, the Hessian can be **indefinite**, producing a saddle surface that is neither convex nor concave.

**Why concavity reduces to convexity:** $f$ is concave iff $-f$ is convex. Thus studying convexity (via the Hessian eigenvalues or PSD/NSD conditions) fully characterizes both convexity and concavity.

## 2. Multivariate Linear Regression as a QP

For OLS with $X\in\mathbb{R}^{n\times p}$ and $y\in\mathbb{R}^n$, we solve

$$\min_{\beta\in\mathbb{R}^p}\ \tfrac{1}{2}\|X\beta - y\|_2^2 = \tfrac{1}{2}\beta^\top (X^\top X)\beta - (X^\top y)^\top\beta + \tfrac{1}{2}y^\top y.$$

This is a quadratic program with $P=X^\top X$ and $q=-X^\top y$.

```python
# Synthetic data
n, p = 200, 4
X = np.random.randn(n, p)
beta_true = np.array([1.5, -2.0, 0.7, 3.0])
noise = 0.5 * np.random.randn(n)
y = X @ beta_true + noise

# QP matrices
P = X.T @ X
q = -(X.T @ y)

# Closed-form OLS (normal equations)
beta_ols = np.linalg.solve(P, X.T @ y)

# QP solve with scipy.optimize
obj = lambda b: 0.5 * b @ P @ b + q @ b
jac = lambda b: P @ b + q

res = minimize(obj, x0=np.zeros(p), jac=jac, method="BFGS")
beta_qp = res.x

print("True beta:", beta_true)
print("OLS beta:", beta_ols.round(4))
print("QP beta: ", beta_qp.round(4))
print("Objective value:", obj(beta_qp).round(4))
```

The QP solution matches the closed-form OLS estimate up to numerical tolerance, as expected for a strictly convex quadratic objective.

## 3. Markowitz Portfolio Optimization as a QP

Given expected returns $\mu$ and covariance $\Sigma$, the classical Markowitz problem is

$$\min_w \tfrac{1}{2} w^\top \Sigma w \quad \text{s.t. } \mathbf{1}^\top w = 1,\ \mu^\top w = r_\text{target},\ w \ge 0.$$

This is a convex QP when $\Sigma$ is positive semidefinite.

```python
# Synthetic Markowitz data
m = 6
A = np.random.randn(m, m)
Sigma = A.T @ A  # PSD covariance
mu = np.random.uniform(0.05, 0.20, size=m)

r_target = 0.12

# Objective: 0.5 * w^T Sigma w
obj = lambda w: 0.5 * w @ Sigma @ w
jac = lambda w: Sigma @ w

# Constraints: sum w = 1, mu^T w = r_target
cons = [
    {"type": "eq", "fun": lambda w: np.sum(w) - 1.0},
    {"type": "eq", "fun": lambda w: mu @ w - r_target},
]

# Bounds: w >= 0
bounds = [(0.0, 1.0) for _ in range(m)]

w0 = np.ones(m) / m
res = minimize(obj, x0=w0, jac=jac, constraints=cons, bounds=bounds, method="SLSQP")
w_star = res.x

print("Expected returns:", mu.round(3))
print("Optimal weights:", w_star.round(4))
print("Portfolio return:", (mu @ w_star).round(4))
print("Portfolio variance:", (w_star @ Sigma @ w_star).round(6))
```

The optimizer returns a feasible weight vector that hits the target return while minimizing variance, which is the classic convex QP interpretation of meanvariance optimization.

## 4. Convex vs. Non-Convex Sets in $\mathbb{R}^2$

A set $S$ is convex if it contains the line segment between any two points in $S$. We illustrate:
- **Convex set:** disk $x^2 + y^2 \le 1$ (sublevel set of a convex quadratic).
- **Non-convex set:** annulus $1 \le x^2 + y^2 \le 4$, which has a hole and fails the line-segment property.

```python
x = np.linspace(-2.5, 2.5, 400)
y = np.linspace(-2.5, 2.5, 400)
X, Y = np.meshgrid(x, y)
R2 = X**2 + Y**2

convex_set = R2 <= 1.0
nonconvex_set = (R2 >= 1.0) & (R2 <= 4.0)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].contourf(X, Y, convex_set, levels=1, colors=["#88ccee"])
ax[0].set_title(r"Convex set: $x^2+y^2\leq 1$")
ax[0].set_aspect('equal')

ax[1].contourf(X, Y, nonconvex_set, levels=1, colors=["#cc6677"])
ax[1].set_title(r"Non-convex set: $1\leq x^2+y^2\leq 4$")
ax[1].set_aspect('equal')

for a in ax:
    a.set_xlabel("x")
    a.set_ylabel("y")

plt.tight_layout()
plt.show()
```

**Relation to function convexity:** if $f$ is convex, every sublevel set $\{x : f(x) \le \alpha\}$ is convex. The disk is a sublevel set of the convex quadratic $x^2+y^2$. The annulus is **not** a sublevel set of any convex function; it is formed by intersecting a convex inequality with a *reverse* (non-convex) inequality, which creates a hole.

## 5. Custom Business QP: Staffing Allocation

**Scenario:** A firm allocates staffing levels $s_i$ across 4 stores. Each store has target demand $d_i$. Understaffing or overstaffing has a quadratic penalty $(s_i-d_i)^2$. Labor cost is linear with rate $c_i$. A total budget $B$ limits spending.

QP formulation:

$$\min_{s\ge 0}\ \sum_i (s_i-d_i)^2 \quad \text{s.t. } c^\top s \le B.$$

```python
# Synthetic staffing data
k = 4
d = np.array([18, 25, 15, 22])           # target staffing
c = np.array([120, 100, 110, 130])        # cost per staff unit
B = 7000                                   # total budget

# Objective
obj = lambda s: np.sum((s - d)**2)
jac = lambda s: 2*(s - d)

# Budget constraint c^T s <= B
cons = [{"type": "ineq", "fun": lambda s: B - c @ s}]

bounds = [(0.0, None) for _ in range(k)]

s0 = d.copy()
res = minimize(obj, x0=s0, jac=jac, constraints=cons, bounds=bounds, method="SLSQP")
s_star = res.x

print("Target demand:", d)
print("Optimal staffing:", s_star.round(2))
print("Total cost:", (c @ s_star).round(2))
```

This convex QP balances service-level fit with a linear budget cap, a common structure in labor planning, marketing allocation, and capacity sizing.

## 6. Spacecraft Acceleration Control (QP)

Let $a_1,\dots,a_4$ be accelerations. The constraints are:

$$v_5 = a_1+a_2+a_3+a_4 = 0,$$

$$x_5 = 3.5a_1 + 2.5a_2 + 1.5a_3 + 0.5a_4 = 100.$$

We minimize the quadratic cost $\sum_{t=1}^4 a_t^2$. This is a convex QP with linear equality constraints.

```python
# Solve min ||a||^2 s.t. A a = b
A = np.array([
    [3.5, 2.5, 1.5, 0.5],
    [1.0, 1.0, 1.0, 1.0],
])
b = np.array([100.0, 0.0])

# KKT system
KKT = np.block([
    [np.eye(4), A.T],
    [A, np.zeros((2, 2))]
])
rhs = np.concatenate([np.zeros(4), b])

sol = np.linalg.solve(KKT, rhs)
a_star = sol[:4]

# Compute trajectory
v = [0.0]
x = [0.0]
for t in range(4):
    v_next = v[-1] + a_star[t]
    x_next = x[-1] + v[-1] + 0.5 * a_star[t]
    v.append(v_next)
    x.append(x_next)

print("Optimal accelerations a*:", a_star.round(4))
print("Positions x:", np.array(x).round(4))
print("Velocities v:", np.array(v).round(4))
print("Final x5, v5:", round(x[-1], 4), round(v[-1], 4))
print("Cost sum a^2:", np.sum(a_star**2).round(4))
```
